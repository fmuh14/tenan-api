### data prepocessing

```python
all_turis_rate = rating
all_turis_rate
```

```python
turis_id.drop(['Rating','Time_Minutes','Coordinate','Lat','Long','Unnamed: 11','Unnamed: 12'],axis=1,inplace=True)
turis_id
```

```python
all_tourism = pd.merge(rating.groupby('Place_Id')['Place_Ratings'].mean(),turis_id,on='Place_Id')
all_tourism # melakukan merge kolom yang dibutuhkan untuk model
```

```python
all_tourism
```

```python
all_tourism.isnull().sum()
```

```python
all_tourism = all_tourism.drop(columns="Price")
```

```python
all_tourism
```

```python
city_mapping = {
    'Jakarta': 1,
    'Yogyakarta': 2,
    'Bandung': 3,
    'Semarang': 4,
    'Surabaya': 5
}
```

```python
all_tourism['id_daerah'] = all_tourism.City.map(city_mapping)
```

```python
all_tourism
```

```python
all_tourism.rename(columns={"Place_Id":"id_wisata"},inplace=True)
all_tourism.rename(columns={"Place_Name":"nama_tempat"},inplace=True)
all_tourism.rename(columns={"Place_Ratings":"rating"},inplace=True)
all_tourism.rename(columns={"Description":"description"},inplace=True)
all_tourism.rename(columns={"Category":"category"},inplace=True)
```

```python
city = all_tourism[["City","id_daerah"]]
city.rename(columns={"City":"city"},inplace=True)
```

```python
# Fungsi untuk membuat link
def create_link(nama_tempat):
    link = 'http://maps.google.com/?q=' + nama_tempat.replace(' ', '+')
    return link
```

```python
# Buat kolom 'link' menggunakan metode apply()
all_tourism['alamat'] = all_tourism['nama_tempat'].apply(create_link)
```
